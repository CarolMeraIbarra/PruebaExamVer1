```python
# Importar graficas
import funcionesGraficas.crear_diagrama_barras_horizontales as diagramaBarra
import funcionesGraficas.crear_grafico_pastel as diagramaPastel
import funcionesGraficas.crear_histograma as diagramaHistograma
import funcionesGraficas.crear_ojiva as diagramaOjiva

# Importar funciones
import funciones.calcular_marcas_clase as marcaClase
import funciones.calcular_ancho_clase as anchoClase
import funciones.calcular_limites as limites
import funciones.calcular_numero_intervalos as numIntervalos
import funciones.calcular_rango as rangoo
import funciones.frecuencia_absoluta as frecAbsoluta
import funciones.frecuencia_acumulada as frecAcumulada
import funciones.frecuencia_relativa as frecRelativa
import funciones.frecuencia_absoluta_pe as frecAbsolutaP
import funciones.imprimir_resultados_corto as resultadosC
import funciones.imprimir_resultados_largo as resultadosL
import funciones.imprimir_tabla_corta as tablaC
import funciones.imprimir_tabla_larga as tablaL
import funciones.obtener_valores_extremos as valoresEx
import funciones.ordenar_datos as ordenarD


# Lista de datos
datos = [250, 790, 470, 510, 500, 330, 300, 750, 580, 740, 420, 260, 
         440, 770, 400, 540, 530, 330, 390, 380, 340, 460, 530, 260, 410, 360, 270, 510, 630, 330]

# Ordenar los datos
datos_ordenados = ordenarD.ordenar_datos(datos)

# Obtener valores extremos
valor_minimo, valor_maximo = valoresEx.obtener_valores_extremos(datos_ordenados)

# Calcular el rango
rango = rangoo.calcular_rango(valor_minimo, valor_maximo)

# Calcular el número de intervalos
numero_intervalos = numIntervalos.calcular_numero_intervalos(datos)

# Calcular el ancho de la clase y el número de intervalos
ancho_clase, numero_intervalos = anchoClase.calcular_ancho_clase(rango, numero_intervalos)

# Calcular los límites de los intervalos
limites_inferiores, limites_superiores = limites.calcular_limites(valor_minimo, ancho_clase, numero_intervalos)

# Calcular las marcas de clase
marcas_clase = marcaClase.calcular_marcas_clase(limites_inferiores, limites_superiores)

# Calcular la frecuencia absoluta
frecuencias = frecAbsoluta.frecuencia_absoluta(datos_ordenados, valor_minimo, ancho_clase, numero_intervalos)

# Calcular la frecuencia relativa
frecuencias_relativas = frecRelativa.frecuencia_relativa(frecuencias)

# Calcular la frecuencia acumulada
frecuencias_acumuladas = frecAcumulada.frecuencia_acumulada(frecuencias_relativas)

# Imprimir los resultados
resultadosL.imprimir_resultados_largo(datos_ordenados, valor_minimo, valor_maximo, rango, numero_intervalos, ancho_clase, frecuencias, frecuencias_relativas, frecuencias_acumuladas)

# Imprimir la tabla
tablaL.imprimir_tabla_larga(numero_intervalos, limites_inferiores, limites_superiores, marcas_clase, frecuencias, frecuencias_relativas, frecuencias_acumuladas)


# Crear el histograma
diagramaHistograma.crear_histograma(marcas_clase, frecuencias, ancho_clase)

# Crear ojiva
diagramaOjiva.crear_ojiva(marcas_clase, frecuencias_acumuladas)
```
